# Prediction using LGBM model

```python
import pickle
model_lgbm = pickle.load(open('tuned_lgbm_model.pkl', 'rb'))
```

```python
df_test_actual[targ_col1] = encoder1.transform(df_test_actual[targ_col1])
df_test_actual[label_col1] = encoder2.transform(df_test_actual[label_col1])
```

```python
model_col = targ_col1 + label_col1 + ['hours_per_week', 'income', 'net_cap_gain','Age']

y_pred_test = (model_lgbm.predict_proba(df_test_actual[model_col]))

dumm_df = pd.DataFrame(y_pred_test, columns = ['prob_0','prob'])
df_test_actual = pd.merge(df_test_actual, dumm_df['prob'], left_index=True, right_index=True)
```

```python
df_test_actual['pred_app']=-1
for i in df_test_actual.index:
    if df_test_actual['prob'][i] > 0.42:
        df_test_actual['pred_app'][i] = 1
    else:
        df_test_actual['pred_app'][i] = 0
```

```python
# Considering weighted average of squares
df_test_actual['net_income'] = (10*df_test_actual['income_age']+80*df_test_actual['net_cap_gain'] +200*df_test_actual['income_ed'] + 110*df_test_actual['income_mar_st'] + 50*df_test_actual['income_rel']+ 540*df_test_actual['income_occ'] +70*df_test_actual['income_workclass'] )

max_net_income = 89240260


a=0.35
df_test_actual['credit'] = (df_test_actual['net_income']*a/(max_net_income)+df_test_actual['prob']*(1-a))*8000
df_test_actual['credit'] = df_test_actual['credit'].apply(lambda x: 1000 if x < 3000 else 8000 if x>5800 else x)
```

```python
# Converting credit to nearest 500
import math

df_test_actual['credit'] = df_test_actual['credit'].apply(lambda x: math.ceil(x / 500.0) * 500.0)
```

```python
df_test_actual['label'] = 1
for j in df_test_actual.index:
    if ((df_test_actual['pred_app'][j] == 0) & (df_test_actual['approved'][j] == 0)):
        df_test_actual['label'][j] = 'TN'

    elif  ((df_test_actual['pred_app'][j] == 1) & (df_test_actual['approved'][j] == 1)):
        df_test_actual['label'][j] = 'TP'

    elif  ((df_test_actual['pred_app'][j] == 1) & (df_test_actual['approved'][j] == 0)):
        df_test_actual['label'][j] = 'FP'

    elif  ((df_test_actual['pred_app'][j] == 0) & (df_test_actual['approved'][j] == 1)):
        df_test_actual['label'][j] = 'FN'

revenue = df_test_actual[df_test_actual['label']=='TP']['credit'].sum()
psuedo_loss = df_test_actual[df_test_actual['label']=='FN']['credit'].sum()
actual_loss = df_test_actual[df_test_actual['label']=='FP']['credit'].sum()
total_credit = df_test_actual['credit'].sum()


print('Revenue :',revenue, revenue/total_credit*100,'\nPsuedo loss:',psuedo_loss, psuedo_loss/total_credit*100,'\nActual loss:', actual_loss, actual_loss/total_credit*100 )
print('profit', revenue-psuedo_loss-actual_loss)
```
